```python
import pandas as pd
import requests
```

```python
#api address 
url = "https://www.balldontlie.io/api/v1/players"
```

```python
#pulls in datat from API
r = requests.get(url)
```

```python
#extracts data from requests as json file
data = r.json()
```

```python
#json file
data
```

```python
#shows keys of json file 
data.keys()
```

```python
#check to see the type of each key - look for list
type(data['data'])
```

```python
#check to what the list is made of - look for list dict
type(data['data'][0])
```

```python
#creates dataframe of json file over specific key
df=pd.DataFrame(data['data'])
```

```python
#select and print the first 4 rows
df.head(4)
```

```python
#select and print the 2nd and 3rd row
df.iloc[2:4]
```

```python
df.shape
```

```python
#find and print descriptive stats
#calculates mean of weight
avg_weight = df['weight_pounds'].mean()
print(avg_weight)
```

```python
#find and print descriptive stats
#calculates median of _id (only int column)
med_weight = df['weight_pounds'].median()
print(med_weight)
```

```python
#write a query
#shows value with first name Jabari
value = 'Jabari'
df.query('first_name == @value')
```

```python
#write a query
#shows rows with weights greater than the avg weight
df.query('weight_pounds > @avg_weight')
```
